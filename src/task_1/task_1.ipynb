```python
import sqlite3 as lite
import pandas as pd

con = lite.connect('./data/checking-logs.sqlite') # подключение БД к Питон

pd.io.sql.read_sql("PRAGMA table_info(pageviews)", con) # PRAGMA получение информации о полях таблицы
```

```python
pd.io.sql.read_sql("""
    SELECT * FROM pageviews 
    LIMIT 10
    """,con) # первые 10 строк таблицы
```

```python
pageviews = pd.io.sql.read_sql("""
    SELECT uid, datetime FROM pageviews
    WHERE uid LIKE 'user_%'
    ORDER BY LENGTH(uid), uid
    """, con, index_col ='datetime', parse_dates = ['datetime']) # parse_dates - преобразует в формат датавремя; index_col - индексирует
pageviews
```
